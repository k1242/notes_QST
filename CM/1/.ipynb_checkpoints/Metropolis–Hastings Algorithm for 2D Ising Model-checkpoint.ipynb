```python
import torch
import torch.nn.functional as F

import numpy as np
from scipy.special import gamma

import matplotlib.pyplot as plt
from matplotlib import cm
import time

from IPython.display import clear_output

plt.rcParams["font.family"] = "serif"
```

```python
# Initializing the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Convolutional filter for 2D Ising with J=1.0
kernel = torch.tensor([[0.0, 1.0, 0.0],
                       [1.0, 0.0, 1.0],
                       [0.0, 1.0, 0.0]], device=device).reshape(1, 1, 3, 3)
```

## base

```python
# Function for performing the Metropolis step on one sublattice
def metropolis_sublattice_step(system, T, mask):
    # Calculating the sum of neighbors via convolution
    neighbor_sum = F.conv2d(system, kernel, padding=1)

    # Energy change calculation
    dE = 2 * system * neighbor_sum

    # Calculating the probability of adoption and updating the system
    acceptance_prob = torch.exp(-dE / T).clamp(max=1)
    random_prob = torch.rand_like(system)
    update_mask = (dE <= 0) | (random_prob < acceptance_prob)
    system[update_mask & mask] *= -1
    
    return system

def calculate_total_energy(system):
    # Calculating the sum of neighbors via convolution
    neighbor_sum = F.conv2d(system, kernel, padding=1, groups=1)

    # Calculation of the total energy of the system
    total_energy = - torch.sum(system * neighbor_sum, axis=(1,2,3)) / 2
    return total_energy

# system initialization
def upd_system(batch_size, L, device):
    return torch.randint(0, 2, (batch_size, 1, L, L), dtype=torch.float32, device=device) * 2 - 1

# Creating masks for subgrids
def upd_mask(system):
    mask_A = torch.zeros_like(system, dtype=torch.bool)
    mask_B = torch.zeros_like(system, dtype=torch.bool)
    mask_A[:, :, ::2, ::2] = 1 # Even rows, even columns
    mask_A[:, :, 1::2, 1::2] = 1  # Odd rows, odd columns
    mask_B = ~mask_A
    return mask_A, mask_B

def ising_evol(T, L, n_steps=100, batch_size=100, system_IC="empty"):
    """
    :param T: system temperature
    :param L: lattice linear size
    :param n_steps: number of equilibration steps before the measurements start
    :param batch_size: number of bins use for the error analysis
    :param system_IC: initial configuration of the grid
    :retrun: (C_mean, C_std)
    :retrun: (E_mean, E_std)
    """
    
    # Initializing the spin system
    if system_IC == "empty":
        system = upd_system(batch_size, L, device)
    else:
        system = system_IC
    mask_A, mask_B = upd_mask(system)
    
    # Running the simulation
    for step in range(n_steps):
        system = metropolis_sublattice_step(system, T, mask_A) # Обновление подрешётки A
        system = metropolis_sublattice_step(system, T, mask_B) # Обновление подрешётки B
    
    return system
```

## c) Nice plots

```python
n_dots = 100 # amount of dots
L = 1000  # lattice size
batch_size = int(1e7/ L**2)

Ts = torch.linspace(0.1, 6.0, n_dots, dtype=torch.float32)
Es = torch.zeros_like(Ts)

t1 = time.time()
ising_evol(T, L, n_steps=100, batch_size=batch_size)
t2 = time.time()

for (j, T) in enumerate(Ts):
    clear_output(wait=True)
    print(f"{j+1:02d} / {n_dots}, T={T:.2f}, wait for {(t2-t1) * (n_dots-j):.2f}s")
    system = ising_evol(T, L, n_steps=100, batch_size=batch_size)
    Es[j] = torch.mean(calculate_total_energy(system)).item() / L**2
clear_output()

Cs_diff = (Es[2:] - Es[:-2]) / (Ts[2] - Ts[0])
Ts_diff = Ts[1:-1]
```

```python
# 2D Ising crytical temperature
Tc = 2 / np.log(1 + np.sqrt(2))
fig, axs = plt.subplots(1, 2, figsize=(6,3))
color = "blue"

axs[0].plot(Ts, Es, color=color, alpha=0.5)
axs[0].scatter(Ts, Es, color=color, s=4)
axs[1].plot(Ts_diff, Cs_diff, color=color, alpha=0.5)
axs[1].scatter(Ts_diff, Cs_diff, color=color, s=4)

axs[0].set_ylabel("Energy per site E")
axs[1].set_ylabel("Specific heat capacity C")
for ax in axs:
    ax.grid(alpha=0.5)
    ax.set_xlabel("Temperature T")
    ax.axvline(Tc, color='r', linestyle='--')
plt.tight_layout()
plt.show()
```

## c) Finite-size scaling: Binder 4th-order Cumulant

```python
# https://courses.physics.illinois.edu/phys466/fa2016/lnotes/PhaseTransitions.pdf
def get_U4(system):
    M = torch.squeeze(torch.mean(system, axis=(2,3)))
    M2 = torch.mean(M**2)
    M4 = torch.mean(M**4)
    return (1 - M4 / 3 / M2**2).item()
```

```python
n_dots = 100 # количество точек на C(T)
Ls = [8, 16, 32]  # размер решёток
colors = ["blue", "purple", "red"]

Ts = torch.linspace(1.0, 4.0, n_dots, dtype=torch.float32)
Us = torch.zeros((len(Ls), n_dots))

fig, axs = plt.subplots(1, 3, figsize=(9,3))
for (i, (L, color)) in enumerate(zip(Ls, colors)):
    batch_size = int(1e7/ L**2)

    Es = torch.zeros_like(Ts)

    t1 = time.time()
    ising_evol(T, L, n_steps=100, batch_size=batch_size)
    t2 = time.time()

    for (j, T) in enumerate(Ts):
        clear_output(wait=True)
        print(f"{L}, {j+1:02d} / {n_dots}, T={T:.2f}, wait for {(t2-t1) * (n_dots-j):.2f}s")
        system = ising_evol(T, L, n_steps=100, batch_size=batch_size)
        Es[j] = torch.mean(calculate_total_energy(system)).item() / L**2
        Us[i, j] = get_U4(system)
    clear_output()

    Cs_diff = (Es[2:] - Es[:-2]) / (Ts[2] - Ts[0])
    Ts_diff = Ts[1:-1]
    
    axs[0].plot(Ts, Es, color=color, alpha=0.75, label=f"L={L}")
#     axs[0].scatter(Ts, Es, color=color, s=4)
    axs[1].plot(Ts_diff, Cs_diff, color=color, alpha=0.75)
#     axs[1].scatter(Ts_diff, Cs_diff, color=color, s=4)

axs[2].plot(Ts, Us[0], color=colors[0], alpha=0.75)
axs[2].plot(Ts, Us[1], color=colors[1], alpha=0.75)
axs[2].plot(Ts, Us[2], color=colors[2], alpha=0.75)

axs[0].legend()
axs[0].set_ylabel("Energy per site E")
axs[0].set_ylim(-2,-0.25)
axs[1].set_ylabel("Specific heat capacity C")
axs[2].set_ylabel("Binder 4th-order Cumulant U4")
for ax in axs:
    ax.grid(alpha=0.5)
    ax.set_xlabel("Temperature T")
    ax.axvline(Tc, color='r', linestyle='--')
plt.tight_layout()
plt.show()
```

## e)

```python
# Функция для выполнения шага Метрополиса на одной подрешётке
def metropolis_sublattice_step_h(system, T, mask):
    # Вычисление суммы соседей через свёртку
    neighbor_sum = F.conv2d(system, kernel, padding=1)

    # Расчёт изменения энергии
    dE = 2 * system * neighbor_sum
    dE[system > 0] += 2 * h
    dE[system < 0] -= 2 * h

    # Расчёт вероятности принятия и обновление системы
    acceptance_prob = torch.exp(-dE / T).clamp(max=1)
    random_prob = torch.rand_like(system)
    update_mask = (dE <= 0) | (random_prob < acceptance_prob)
    system[update_mask & mask] *= -1
    
    return system

def ising_evol_h(T, L, n_steps=100, batch_size=100, system_IC="empty"):
    """
    :param T: температура системы
    :param L: размер решётки
    :param n_steps: number of equilibration steps before the measurements start
    :param batch_size: number of bins use for the error analysis
    :retrun: (C_mean, C_std)
    :retrun: (E_mean, E_std)
    """
    
    # Инициализация системы спинов
    if system_IC == "empty":
        system = upd_system(batch_size, L, device)
    else:
        system = system_IC
    mask_A, mask_B = upd_mask(system)
    
    # Запуск симуляции
    for step in range(n_steps):
        system = metropolis_sublattice_step_h(system, T, mask_A) # Обновление подрешётки A
        system = metropolis_sublattice_step_h(system, T, mask_B) # Обновление подрешётки B
    
    return  system
```

```python
h = 0.05
n_dots = 50 # количество точек на C(T)
L = 50  # Размер решётки
batch_size = int(1e6/ L**2)

Ts = torch.linspace(0.1, 6.0, n_dots, dtype=torch.float32)
Es = torch.zeros_like(Ts)
Ms = torch.zeros_like(Ts)

t1 = time.time()
ising_evol(T, L, n_steps=1000, batch_size=batch_size)
t2 = time.time()

for (j, T) in enumerate(Ts):
    clear_output(wait=True)
    print(f"{j+1:02d} / {n_dots}, T={T:.2f}, wait for {(t2-t1) * (n_dots-j):.2f}s")
    system = ising_evol_h(T, L, n_steps=1000, batch_size=batch_size)
    Es[j] = torch.mean(calculate_total_energy(system)).item() / L**2
    Ms[j] = torch.mean(system).item()
clear_output()
```

```python
Tc = 2 / np.log(1 + np.sqrt(2))
fig, axs = plt.subplots(1, 2, figsize=(6,3))
color = "blue"

axs[0].plot(Ts, Es, color=color, alpha=0.5)
axs[0].scatter(Ts, Es, color=color, s=4)
axs[1].plot(Ts, Ms, color=color, alpha=0.5)
axs[1].scatter(Ts, Ms, color=color, s=4)

axs[0].set_ylabel("Energy per site E")
axs[1].set_ylabel("average M")
for ax in axs:
    ax.grid(alpha=0.5)
    ax.set_xlabel("Temperature T")
    ax.axvline(Tc, color='r', linestyle='--')
plt.tight_layout()
plt.show()
```
