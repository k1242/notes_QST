## Exercise 4: Exact diagonalization (PBC)

```python
# a)
Id = sp.sparse.csr_matrix([[1,0],[0,1]]);
Sx = sp.sparse.csr_matrix([[0,1],[1,0]]);
Sz = sp.sparse.csr_matrix([[1,0],[0,-1]]);
```

```python
# b, c)
def get_S_operator(sigma, j, L):
    """
    :param sigma: local operator
    :param j: number of the lattice site
    :param L: size of the system
    :return Sj: 2^L * 2^L full space operator matrix
    """
    operator_base = [sp.sparse.csr_matrix([[1,0],[0,1]]) for _ in range(L)];
    operator_base[j] = sigma
    Sj = sp.sparse.kron(operator_base[0], operator_base[1], format="csr")
    for i in range(2,L):
        Sj = sp.sparse.kron(Sj, operator_base[i], format="csr")
    return Sj

def get_S_operator_list(sigma, L):
    """
    :param sigma: local operator
    :param L: size of the system
    :return: list of full space operators
    """
    return [get_S_operator(sigma, j, L) for j in range(L)]
```

```python
# d)
def gen_hamiltonian(sx_list, sz_list, g, J=1.0, pbc_flag=True):
    """
    :param sx_list: list of full space Sx operators
    :param sz_list: list of full space Sz operators
    :param g: magnetic field
    :param J: hopping coefficient
    :return H: the transverse field Ising model in 1D Hamiltonian
    """
    L = len(sx_list); #size of the system
    H = sp.sparse.csr_matrix((2**L,2**L));
    for (O1, O2) in zip(sx_list[1:], sx_list[:-1]):
        H += (-J) * O1 * O2
    if pbc_flag:
        H += (-J) *  sx_list[-1] * sx_list[0] #periodic boundary conditions
    for O in sz_list:
        H += (-g) * O
    return H

L=2;
H = gen_hamiltonian(get_S_operator_list(Sx, L), get_S_operator_list(Sz, L), 0.1);
H.toarray()
```

```python
# e
L=2;
H = gen_hamiltonian(get_S_operator_list(Sx, L), get_S_operator_list(Sz, L), 0.1);
evals, evecs = sp.sparse.linalg.eigsh(H, 3, which="SA");
evals
```

```python
# compare sparse eigsh and full eigh
L = 12;
H = gen_hamiltonian(get_S_operator_list(Sx, L), get_S_operator_list(Sz, L), 0.1);
t1 = time.time()
evals, evecs = sp.sparse.linalg.eigsh(H, 3, which="SA");
t2 = time.time()
evals_full, evecs_full = sp.linalg.eigh(H.toarray());
t3 = time.time()

print(f"sparse eigsh time: {t2-t1:.2f}s")
print(f"eigh time: {t3-t2:.1f}s")
```

```python
# f
def get_TFIM_ground_state(L, g, J=1.0, n=1, pbc_flag=True):
    """
    :param L: size of the system
    :param g: magnetic field
    :param J: hopping coefficient
    :param n: amount of calculating lowest energy states
    :return: ground energy, ground state
    """
    H = gen_hamiltonian(get_S_operator_list(Sx, L), get_S_operator_list(Sz, L), g, J, pbc_flag);
    eval, evec = sp.sparse.linalg.eigsh(H, n, which="SA");
    return eval, evec


def get_largest_distance_ss_corr(L, g, pbc_flag=True):
    """
    :param L: size of the system
    :param g: magnetic field
    :return: the largest-distance spin-spin correlation
    """
    _, state = get_TFIM_ground_state(L, g);
    O = get_S_operator(Sx, 0, L) * get_S_operator(Sx, L//2, L)
    return np.inner(np.conjugate(state).flatten(), (O * state).flatten())

plt.figure(figsize=(3,3))
for L in [6, 8, 10, 12]:
    print(f"{L=} calculating")
    gs = np.linspace(0,2,20);
    Cs = [get_largest_distance_ss_corr(L, g) for g in gs];
    clear_output(wait=True)
    plt.plot(gs,Cs, label=f"L={L}")

plt.xlabel("g, magnetic field")
plt.ylabel("C, largest-distance spin-spin corr")
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```

```python
# g
L = 12;

l = 20;
gs = np.linspace(0,2,l);
E10,E20 = np.zeros((2,l));
for (i,g) in enumerate(gs):
    (E0, E1, E2), _ = get_TFIM_ground_state(L, g, n=3);
    E10[i] = E1-E0
    E20[i] = E2-E0

plt.figure(figsize=(3,3))
plt.plot(gs,E10, label="E1-E0", color="red")
plt.plot(gs,E20, label="E2-E0", color="blue")
plt.xlabel("g, magnetic field")
plt.ylabel("the excitation energies gap")
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

The result coincide with expectations, because we have closeg gap in the phase transition point and ground state FM degeneracy = 2 at g < 1.0.

### Finite system DMRG

Next code based on article [arXiv:2304.13395](https://arxiv.org/abs/2304.13395).

```python
I = np.array([[1,0],[0,1]]);
Sx = np.array([[0,1],[1,0]]);
Sz = np.array([[1,0],[0,-1]]);
```

```python
def gen_hamiltonian_dmrg(Nit, g, J=1.0, D=14, save_flag=False, tol_deg=1e-10):
    """
    :param Nit: numberr of iterations, Nspins = 2+2*Nit
    ...
    :param tol_deg: tolerance for degeneracies
    :retrun if save_flag: H_SBs for L=2,...,2+2*Nit
    :retrun if not save_flag: H_SB for L=2+2*Nit
    """
    O_L_list = []
    O_R_list = []

    if save_flag:
        H_SBs = []
    
    # truncated iterative diagonalization
    start_time = time.time()
    for k in range(1, Nit+1):
        if k==1: 
            I_S = I
            H_S = - g * Sz
            Sxborder_S = Sx
    
            I_E = I
            H_E = - g * Sz
            Sxborder_E = Sx
        else:
            # truncation by lowest-eigenvalues
            En_L, psin_L = np.linalg.eigh(H_L)
            O_L = psin_L[:,0]
            for it in range(1, min(D, len(En_L))):
                O_L = np.vstack([O_L, psin_L[:,it]])
    
            # degenerate states processing
            lines_O_L = np.shape(O_L)[0]
            for it in range(lines_O_L,len(En_L)):
                if abs(En_L[lines_O_L-1] - En_L[it]) < tol_deg:
                    O_L = np.vstack([O_L, psin_L[:,it]])
                else:
                    break
            # O_R
            En_R, psin_R = np.linalg.eigh(H_R)
            O_R = psin_R[:,0]
            for it in range(1, min(D,len(En_R))):
                O_R = np.vstack([O_R, psin_R[:,it]])
            
            # degenerate states processing
            lines_O_R = np.shape(O_R)[0]
            for it in range(lines_O_R, len(En_R)):
                if abs(En_R[lines_O_R-1] - En_R[it]) < tol_deg:
                    O_R = np.vstack([O_R, psin_R[:,it]])
                else:
                    break
    
            # tracked O
            O_L_list.append(O_L)
            O_R_list.append(O_R)
            
            # S
            I_S = np.matmul(np.matmul(O_L,I_L),O_L.conj().T)
            H_S = np.matmul(np.matmul(O_L,H_L),O_L.conj().T)
            Sxborder_S = np.matmul(np.matmul(O_L,Sxborder_L),O_L.conj().T)
            
            # E
            I_E = np.matmul(np.matmul(O_R,I_R),O_R.conj().T)
            H_E = np.matmul(np.matmul(O_R,H_R),O_R.conj().T)
            Sxborder_E = np.matmul(np.matmul(O_R,Sxborder_R),O_R.conj().T)
    
        # H_L
        I_L = np.kron(I_S,I)
        H_L = np.kron(H_S,I) + (-g) * np.kron(I_S, Sz) #H_S x I_l
        H_L += (-J) * np.kron(Sxborder_S,Sx)
        Sxborder_L = np.kron(I_S,Sx)
    
        # H_R
        I_R = np.kron(I, I_E)
        H_R = np.kron(I, H_E) + (-g) * np.kron(Sz, I_E) #I_r x H_E
        H_R += (-J) * np.kron(Sx, Sxborder_E)
        Sxborder_R = np.kron(Sx, I_E)

        # H_SB
        if save_flag:
            H_SBs.append(H_SB)

    if save_flag:
        return H_SBs, O_L_list, O_R_list
    else:
        H_SB = np.kron(H_L, I_R) #H_L x I_R
        H_SB += np.kron(I_L, H_R) #I_L x H_R
        H_SB += np.kron(Sxborder_L, Sxborder_R)
    
        return H_SB, O_L_list, O_R_list
```

```python
def get_S_operator_list_dmrg(O_L_list, O_R_list, S):
    """
    :param O_L_list: list of O_L opeartors during dmrg iterations
    :param O_R_list: list of O_R opeartors during dmrg iterations
    :param S: single site operator
    :return Sj: list of full space operators
    """
    # L
    Sj_L = [np.kron(S,I),np.kron(I,S)]
    for it in range(len(O_L_list)):
        for it2 in range(len(Sj_L)):
            Sj_L[it2] = np.kron( np.matmul(np.matmul(O_L_list[it],Sj_L[it2]),O_L_list[it].conj().T) , I)
        Sj_L.append(np.kron( np.matmul(O_L_list[it],O_L_list[it].conj().T) , S ))
    
    # R
    Sj_R = [np.kron(S,I),np.kron(I,S)]
    for it in range(len(O_R_list)):
        for it2 in range(len(Sj_R)):
            Sj_R[it2] = np.kron(I, np.matmul(np.matmul(O_R_list[it],Sj_R[it2]),O_R_list[it].conj().T))
        Sj_R.insert(0,np.kron(S, np.matmul(O_R_list[it],O_R_list[it].conj().T)))
    
    # SB
    Sj = []
    I_R = np.kron( I , np.matmul(O_R_list[-1],O_R_list[-1].conj().T) )
    I_L = np.kron( np.matmul(O_L_list[-1],O_L_list[-1].conj().T) , I )
    for it in range(len(Sj_L)):
        Sj.append( np.kron(Sj_L[it],I_R) )
    for it in range(len(Sj_R)):
        Sj.append( np.kron(I_L,Sj_R[it]) )

    return Sj
```

```python
def get_TFIM_ground_state_dmrg(L, g, J=1.0, n=1):
    """
    :param L: size of the system
    :param g: magnetic field
    :param J: hopping coefficient
    :return: ground energy, ground state for n=1 and pbc_flag=False
    """
    H_SB, O_L_list, O_R_list = gen_hamiltonian_dmrg(L//2-1, g);
    En, psin = np.linalg.eigh(H_SB);
    return En[:n], psin[:, :n], O_L_list, O_R_list

def get_largest_distance_ss_corr_dmrg(L, g, J=1.0):
    """
    :param L: size of the system
    :param g: magnetic field
    :return: the largest-distance spin-spin correlation
    """
    E0, state, O_L_list, O_R_list = get_TFIM_ground_state_dmrg(L, g, J);
    if len(O_L_list) > 0:
        Sj = get_S_operator_list_dmrg(O_L_list, O_R_list, Sx);
    else:
        Sj = get_S_operator_list(Sx, L);
        Sj = [S.toarray() for S in Sj];
    return np.matmul(np.matmul(state.conj().T, np.dot(Sj[0], Sj[L//2])), state).item()
```
