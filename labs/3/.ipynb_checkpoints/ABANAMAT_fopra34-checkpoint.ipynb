## Base

```python
def add_text(ax, text, x, y, alpha_x, alpha_y):
    """
    Adds text to an Axes object at a calculated position based on data ranges and relative positions.

    Args:
    ax (matplotlib.axes.Axes): Target Axes for the text.
    text (str): Text to be added.
    x (array-like): X data to determine positioning.
    y (array-like): Y data to determine positioning.
    alpha_x (float): Relative horizontal position within the data range (0 to 1).
    alpha_y (float): Relative vertical position within the data range (0 to 1).
    
    This function positions text relative to the min and max of provided x and y data,
    adjusted by alpha_x and alpha_y, which scale the text's position within the data bounds.
    """
    # Calculate text position from data bounds and relative positioning
    x_pos = np.min(x) + alpha_x * (np.max(x) - np.min(x))
    y_pos = np.min(y) + alpha_y * (np.max(y) - np.min(y))

    # Add text at calculated position with predefined style
    ax.text(
        x_pos, y_pos, text, fontsize=10, color='black',
        bbox={'facecolor': 'white', 'alpha': 0.5, 'pad': 4, 'edgecolor': 'black'}
    )
```

# 4.2 Rabi Oscillation

Consider the system with Hamiltonian
$$\hat{H} = - \frac{\omega_0}{2} \hat{\sigma}_z +\omega_1 \cos(\omega t) \hat{\sigma}_x$$
The dynamics of state $\ket{\psi} = \alpha(t) \ket{0} + \beta(t) \ket{1}$is given by 
$$ e^{-i\omega t/2} \alpha(t) = \cos\left(\frac{\Omega t}{2}\right) - i \frac{\Delta}{\Omega} \sin\left(\frac{\Omega t}{2}\right)$$$$e^{i\omega t/2} \beta (t) = -i \frac{\omega_1}{\Omega} \sin\left(\frac{\Omega t}{2}\right)$$
with $\Omega = \sqrt{\omega_1^2 + \Delta^2}$ and $\Delta = \omega - \omega_0$.$$

## Exercise 5.

The time-dependent population for states initialized with $\alpha(0) = 1$:
$$P_0(t) = \frac{\Delta ^2 }{\Omega ^2}\sin ^2\left(\frac{t \Omega }{2}\right)+\cos ^2\left(\frac{t \Omega }{2}\right)$$
$$P_1(t) = \frac{\omega_1^2 }{\Omega^2}\sin ^2\left(\frac{t \Omega }{2}\right)$$

## Exercise 6.

```python
def time_evolve(circ, t, dt, w, w0, w1):
    """
    Apply the time-evolution operator U(t) using Trotterization.
    Arguments:
    - circ: QuantumCircuit, the circuit to which the evolution is applied
    - t: float, total time of the evolution
    - dt: float, time step for the Trotterization
    - w: float, frequency of the external field
    - w0: float, system frequency
    - w1: float, strength of the driving field
    """
    N = int(t / dt)  # number of time steps
    for n in range(N):
        circ.rz(w0 * dt, 0)  # Rotation around Z-axis
        circ.rx(-w1 * dt * np.cos(w * dt * n) * 2, 0)  # Rotation around X-axis, modulated by cos(w*dt)

    return circ
```

```python
def do_exp_6(t, w = 25.0, w0 = 25.0, w1 = 2.0, n_shots=400, dt = 0.05):
    """
    dt -- time step
    w  -- external frequency
    w0 -- system frequency
    w1 -- driving field strength
    """
    # Create a new quantum circuit with one qubit
    qc = QuantumCircuit(1)
    
    # Time evolve the circuit
    time_evolve(qc, t, dt, w, w0, w1)
    
    # Add measurement
    qc.measure_all()
    
    # Simulate the circuit using the recommended API
    # simulator = Aer.get_backend('statevector_simulator') #result.get_statevector()
    simulator = Aer.get_backend('qasm_simulator')
    transpiled_circuit = transpile(qc, simulator)
    job = simulator.run(transpiled_circuit, shots=n_shots)
    result = job.result()
    counts = result.get_counts()

    return counts.get('1', 0) / n_shots
```

```python
ts = np.arange(0.05, 4, 0.05)
Ps_a = [do_exp_6(t, w=25.5, w0=25.0, w1=2.0) for t in ts]
```

```python
ws = np.arange(10, 40, 0.25)
Ps_b = [do_exp_6(np.pi / 2.0, w=w, w0=25.0, w1=2.0) for w in ws]
```

```python
ts = np.arange(0.05, 4, 0.05)
Ps_c = [do_exp_6(t, w=2.0, w0=2.0, w1=2.0) for t in ts]
```

```python
fig, [ax_a, ax_b, ax_c] = plt.subplots(1, 3, figsize=(9,3))

Omega = np.sqrt(2**2 + (0.5)**2)
ax_a.plot(ts, Ps_a, "blue")
ax_a.plot(ts, (2 / Omega * np.sin(Omega * ts / 2))**2, "--k")
ax_a.set_title("a) Rabi Oscillation")
ax_a.set_xlabel("t")
ax_a.set_ylabel("P1")
add_text(ax_a, "w0=25.0\nw=25.5\nw1=2.0", ts, Ps_a, 0.7, 0.75)

ax_b.plot(ws, Ps_b, "blue")
ax_b.set_title("b) Resonance")
ax_b.set_xlabel("w")
ax_b.set_ylabel("P1")
add_text(ax_b, "t=pi/2\nw0=25\nw1=2.0", ws, Ps_b, 0.75, 0.75)

ax_c.plot(ts, Ps_c, "blue")
ax_c.plot(ts, (np.sin(ts))**2, "--k")
ax_c.set_title("c) RWA broken")
ax_c.set_xlabel("t")
ax_c.set_ylabel("P1")
add_text(ax_c, "w0=2.0\nw=2.0\nw1=2.0", ts, Ps_c, 0.75, 0.75)

for ax in [ax_a, ax_b, ax_c]:
    ax.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

We see Rabi oscillations in graph a, resonance in graph b, and RWA violation in graph c.

## Exercise 7.

```python
def do_exp_7(t, w=25.5, w0=25.0, w1=2.0, n_shots=400, dt=0.05, gamma=0.02, p=0.1):
    """
    Perform the experiment with noise.
    """
    # Create a new quantum circuit with one qubit
    qc = QuantumCircuit(1)
    
    # Time evolve the circuit
    time_evolve(qc, t, dt, w, w0, w1)

    # Add measurement
    qc.measure_all()
    
    # Define the noise model
    noise_model = NoiseModel()
    ad_error = amplitude_damping_error(gamma, p)
    noise_model.add_all_qubit_quantum_error(ad_error, ['rx', 'rz'])

    # Setup the simulator with the noise model
    simulator = Aer.get_backend('aer_simulator')
    # transpiled_circuit = transpile(qc, simulator)

    # Run the simulation with noise model
    job = simulator.run(qc, noise_model=noise_model, shots=n_shots)
    result = job.result()
    counts = result.get_counts()

    return counts.get('1', 0) / n_shots
```

```python
ts = np.arange(0.05, 10, 0.05)
Ps71 = [do_exp_7(t, gamma=0.002, p=0.1) for t in ts]
Ps72 = [do_exp_7(t, gamma=0.002, p=0.1, dt=0.01) for t in ts]
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(6,3))

Omega = np.sqrt(2**2 + (0.5)**2)
ax1.plot(ts, Ps71, "blue")
ax1.plot(ts, (2 / Omega * np.sin(Omega * ts / 2))**2, "--k")
ax1.set_title("Rabi Oscillation with errors\ndt=0.05")
ax1.set_xlabel("t")
ax1.set_ylabel("P1")
# add_text(ax1, "p=0.1\ngamma=0.002", ts, Ps_a, 0.01, 0.01)

ax2.plot(ts, Ps72, "blue")
ax2.plot(ts, (2 / Omega * np.sin(Omega * ts / 2))**2, "--k")
ax2.set_title("Rabi Oscillation with errors\ndt=0.01")
ax2.set_xlabel("t")
ax2.set_ylabel("P1")

for ax in [ax1, ax2]:
    ax.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

We use operators more often, so gate error has a greater impact than error due to approximation.

## Exercise 8.

```python
from qiskit.providers.fake_provider import Fake5QV1
```

```python
def do_exp_8(t, w=25.5, w0=25.0, w1=2.0, n_shots=400, dt=0.05):
    """
    Perform the experiment with noise.
    """
    # Create a new quantum circuit with one qubit
    qc = QuantumCircuit(1)
    
    # Time evolve the circuit
    time_evolve(qc, t, dt, w, w0, w1)

    # Add measurement
    qc.measure_all()
    
    # Setup the simulator with the noise model
    # simulator = Aer.get_backend('aer_simulator')
    fake_backend = Fake5QV1()
    backend_simulator = AerSimulator.from_backend(fake_backend)
    
    # Specify the basis gates to use in transpilation
    basis_gates = backend_simulator.configuration().basis_gates
    
    # Transpile the circuit for the backend with explicit basis gates
    transpiled_circuit = transpile(qc, backend_simulator, basis_gates=basis_gates)
    
    # Run the simulation with noise model
    job = backend_simulator.run(transpiled_circuit, shots=n_shots)
    result = job.result()
    counts = result.get_counts()

    return counts.get('1', 0) / n_shots
```

```python
ts = np.arange(0.05, 10, 0.05)
Ps8 = [do_exp_8(t) for t in ts]
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(3,3))

Omega = np.sqrt(2**2 + (0.5)**2)
ax1.plot(ts, Ps8, "blue")
ax1.plot(ts, (2 / Omega * np.sin(Omega * ts / 2))**2, "--k")
ax1.set_title("Rabi Oscillation with Fake5QV1\ndt=0.05")
ax1.set_xlabel("t")
ax1.set_ylabel("P1")
ax1.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

At a minimum, we can guarantee that the error there is lower than that simulated in the previous ex.
